# How to save figure(plots and charts)?

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
import random
from matplotlib import style
```

```python
plt.pie([40,30,20])
plt.savefig("pie_chart")
plt.show()
```

```python
plt.pie([40,30,20])
plt.savefig("pie_chart",dpi = 1000 , quality = 99, facecolor = 'g')
plt.show()
```

```python
plt.pie([40,30,20])
plt.savefig("pie_chart1",dpi = 100 , quality = 99, facecolor = 'g')
plt.show()
```
