```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
dfgoogle_playstoreapps = pd.read_csv("C:\\Users\IT's MRK\Desktop\Learning\Matplotlib\googleplaystore.csv", nrows = 1000)
dfgoogle_playstoreapps.shape
```

```python
dfgoogle_playstoreapps
```

```python
x = dfgoogle_playstoreapps["Rating"]
y = dfgoogle_playstoreapps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.scatter(x,y,c ='r')
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r')
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r',marker = "*",s = 100)
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r',marker = "*",s = 100,alpha = 0.5, linewidths = 10)
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r',marker = "*",s = 100,alpha = 0.5, linewidths = 10,edgecolor = 'g')
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r',marker = "*",s = 100,alpha = 0.5, linewidths = 10,edgecolor = 'g',verts = '<')
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y,c ='r',s = 100,alpha = 0.5, linewidths = 10,edgecolor = 'g',verts = '<')
plt.title("Google play store apps",fontsize = 20)
plt.xlabel("Ratings",fontsize =13)
plt.ylabel("Reviews",fontsize = 13)
```
