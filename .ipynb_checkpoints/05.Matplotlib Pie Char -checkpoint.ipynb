# Pie Char 

```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
plt.pie(cls1_std,labels = classes)

plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0,0,0,0,0]
plt.pie(cls1_std,labels = classes,explode =explode )

plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0,0,0.2,0,0]
plt.pie(cls1_std,labels = classes,explode =explode )

plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0.3,0,0.2,0,0]
plt.pie(cls1_std,labels = classes,explode =explode )

plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0.3,0,0.2,0,0]
colors = ['c','b','y','g','r']
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors )
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0.3,0,0.2,0,0]
colors = ['c','b','y','g','r']
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors, autopct = "%0.2f%%",shadow = True )
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0.3,0,0.2,0,0]
colors = ['c','b','y','g','r']
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors, autopct = "%0.2f%%",shadow = True ,radius = 1.4)
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0,0,0.2,0,0]
colors = ['c','b','y','g','r']
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors, autopct = "%0.2f%%",shadow = True ,radius = 1.4,
       startangle = 90)
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0,0,0.2,0,0]
colors = ['c','b','y','g','r']
textprops = {"fontsize":15}
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors, autopct = "%0.2f%%",shadow = True ,radius = 1.4,
       startangle = 90,textprops = textprops)
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artiical Intelligence","Data Science"]
cls1_std = [45,15,35,25,30]
explode = [0,0,0.2,0,0]
colors = ['c','b','y','g','r']
textprops = {"fontsize":15}
plt.pie(cls1_std,labels = classes,explode =explode,colors = colors, autopct = "%0.2f%%",shadow = True ,radius = 1.4,
       startangle = 90,textprops = textprops)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize = (8,5))
colors = [ 'r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w' ]
labels = np.ones(20)

plt.pie([1], colors = 'k',radius = 2.05)
plt.pie(labels, colors = colors, radius = 2.0)

plt.pie([1], colors = 'g',radius = 1.8)
plt.pie([1], colors = 'y',radius = 1.6)
plt.pie([1], colors = 'c',radius = 1.3)
plt.pie([1], colors = 'b',radius = 1.1)
plt.pie([1], colors = 'm',radius = 0.9)

plt.pie([1], colors = 'b',radius = 0.31)
plt.pie(labels, colors = colors, radius = 0.3)

plt.pie([1], colors = 'w',radius = 0.2)
plt.pie([1], colors = 'k',radius = 0.1) 
plt.show()
```
