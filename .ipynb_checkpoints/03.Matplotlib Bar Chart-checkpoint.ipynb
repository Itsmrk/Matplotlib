# Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style 
```

```python
classes = ["Python","R","AI","ML","DS"]
class1_std = [30,10,20,25,10]
class2_std = [40,5,20,20,10]
class3_std = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_std)
```

```python
plt.barh(classes,class1_std)
```

```python
plt.bar(classes,class1_std,width = 0.2)
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge")
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y")
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y",edgecolor = "m")
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5)
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5, alpha = 0.9)
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5, alpha = 0.9,linestyle= "--" )
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "y",edgecolor = "m",linewidth = 5,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "k",edgecolor = "y",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
```

```python
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "k",edgecolor = "y",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students", visible = False)
plt.legend()
```

```python
style.use("ggplot")
plt.bar(classes,class1_std,width = 0.6,align = "edge",color = "k",edgecolor = "m",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std,width = 0.6,align = "edge",color = "k",edgecolor = "m",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class1_std,align = "edge",color = "k",edgecolor = "m",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std,width = 0.6,align = "edge",color = "k",edgecolor = "m",linewidth = 2,
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.title("Bar Chart of Classes " ,fontsize = 15)
plt.xlabel("Classes",fontsize = 12)
plt.ylabel("Students",fontsize = 12)
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_std,width = 0.2,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes,class2_std,width = 0.2,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.title("Bar Chart of Classes " ,fontsize = 15)
plt.xlabel("Classes",fontsize = 12)
plt.ylabel("Students",fontsize = 12)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.bar(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes_index,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.title("Bar Chart of Classes " ,fontsize = 15)
plt.xlabel("Classes",fontsize = 12)
plt.ylabel("Students",fontsize = 12)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.bar(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes_index + width,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.title("Bar Chart of Classes " ,fontsize = 15)
plt.xlabel("Classes",fontsize = 12)
plt.ylabel("Students",fontsize = 12)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.bar(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes_index + width,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.xticks(classes_index+ width, classes)
plt.title("Bar Chart of Classes " ,fontsize = 25)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Students",fontsize = 15)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.bar(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes_index + width,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.legend()
plt.xticks(classes_index+ width, classes,rotation = 20)
plt.title("Bar Chart of Classes " ,fontsize = 25)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Students",fontsize = 15)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.bar(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.bar(classes_index + width,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 2 students" )
plt.bar(classes_index + width + width,class3_std,width,align = "edge",color = "k",
        alpha = 0.9,linestyle= "--",label = "Class 3 students" )
plt.legend()
plt.xticks(classes_index+ width, classes,rotation = 20)
plt.title("Bar Chart of Classes " ,fontsize = 25)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Students",fontsize = 15)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
classes_index = np.arange(len(classes))
plt.barh(classes_index,class1_std,width,align = "edge",color = "b",
        alpha = 0.9,linestyle= "--",label = "Class 1 students" )
plt.barh(classes_index + width,class2_std,width,align = "edge",color = "g",
        alpha = 0.9,linestyle= "--",label = "Class 2 students" )
plt.barh(classes_index + width + width,class3_std,width,align = "edge",color = "k",
        alpha = 0.9,linestyle= "--",label = "Class 3 students" )
plt.legend()
plt.yticks(classes_index+ width, classes,rotation = 20)
plt.title("Bar Chart of Classes " ,fontsize = 25)
plt.ylabel("Classes",fontsize = 15)
plt.xlabel("Students",fontsize = 15)
```
