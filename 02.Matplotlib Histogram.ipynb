# Ploting Histogram

```python
import matplotlib.pyplot as plt
import numpy as np 
import random 
```

```python
ml_std_age = np.random.randint(20,45, (100))
py_std_age = np.random.randint(15,40, (100))
```

```python
ml_std_age
```

```python
py_std_age
```

```python
plt.hist(ml_std_age)
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins)
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8)
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'step')
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'stepfilled')
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'barstacked')
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal")
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal",color ='m')
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal",color ='r',label = "ML Studnet age")
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal",color ='g',label = "ML Studnet age")
plt.hist(py_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal",color ='r',label = "Py Studnet age")
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend(loc = 4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "horizontal",color ='g',label = "ML Studnet age")
plt.hist(py_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "vertical",color ='r',label = "Py Studnet age")
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend(loc = 4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_std_age,py_std_age],bins,rwidth = 0.8, histtype = 'bar',orientation = "vertical",
         color =['g','r'],label = ["ML Studnet age","PY Studnet age"])
#plt.hist(py_std_age,bins,rwidth = 0.8, histtype = 'bar',orientation = "vertical",color ='r',label = "Py Studnet age")
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend(loc = 0)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_std_age,py_std_age],bins,rwidth = 0.8, histtype = 'bar',orientation = "vertical",
         color =['g','r'],label = ["ML Studnet age","PY Studnet age"])
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend(loc = 0)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_std_age,py_std_age],bins,rwidth = 0.8, histtype = 'bar',orientation = "vertical",
         color =['g','r'],label = ["ML Studnet age","PY Studnet age"])
plt.title("ML students Age",fontsize = 15)
plt.xlabel("Students age category",fontsize = 12)
plt.ylabel("No. of Students Age",fontsize = 12)
plt.legend(loc = 0)
plt.show()
```
